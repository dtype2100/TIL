```python
# webkr(웹문서), blog(블로그), cafearticle(카페), kin(지식인)

start, num = 1, 0
current_datetime = datetime.datetime(2020, 3, 6)

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.column_dimensions['B'].width = 100
excel_sheet.column_dimensions['C'].width = 100
excel_sheet.append(['날짜', '제목', '링크'])

for index in range(30):
    encText = urllib.parse.quote("삼성전자")
    # start_number = start + (index * 100)
    targetDt = current_datetime.strftime('%Y%m%d') #20200306
    naver_open_api = 'https://openapi.naver.com/v1/search/blog?query=' + targetDt + encText  # str(start_number)
    header_params = {"X-Naver-Client-Id":client_id, "X-Naver-Client-Secret":client_secret}
    res = requests.get(naver_open_api, headers=header_params)

    if res.status_code == 200:
        data = res.json()
        for item in data['items']:
            num += 1
            excel_sheet.append([targetDt, item['title'], item['link']])
            print(targetDt, item['title'], item['link'])

        current_datetime = current_datetime + datetime.timedelta(days=1)
    else:
        print ("Error Code:", res.status_code)

excel_file.save('naver_blog.xlsx')
excel_file.close()
```
