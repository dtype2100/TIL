```python
 import numpy as np
import pandas as pd

import seaborn as sns
sns.__version__
```

```python
flights = sns.load_dataset('flights')
flights
```

```python
sns.relplot(x='year', y='passengers', kind='line', data=flights);
```

```python
sns.relplot(x='year', y='passengers', kind='line', ci=None, data=flights);
```

```python
sns.relplot(x='year', y='passengers', kind='line', ci='sd',  data=flights);
```

```python
sns.relplot(x='year', y='passengers', kind='line', estimator=None, data=flights);
```

```python
dots = sns.load_dataset('dots')
dots
```

```python
sns.relplot(x='time', y='firing_rate', kind='line', hue='align',  data=dots);
```

```python
sns.relplot(x='time', y='firing_rate', kind='line', hue='align', style='choice',  data=dots);
```

```python
sns.relplot(x='time', y='firing_rate', kind='line', hue='align', style='choice', markers=True, data=dots);
```

```python
sns.relplot(x='time', y='firing_rate', kind='line', hue='align', col='choice', data=dots);
```

```python
sns.relplot(x='time', y='firing_rate', kind='line', hue='align', row='choice', data=dots);
```

```python
fmri = sns.load_dataset("fmri")
fmri
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='subject', style="region", size='event', data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='subject', col="region", size='event', data=fmri);
```

```python
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', kind='line', hue='subject', col="region", size='event', 
            palette=palette, data=fmri);
```

```python
palette = sns.cubehelix_palette(n_colors=14, light=0.8)
sns.relplot(x='timepoint', y='signal', kind='line', hue='subject', col="region", row='event', 
            palette=palette, data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='event', style="event", col='subject', 
            col_wrap=4,
           data=fmri);
```

```python
sns.relplot(x='timepoint', y='signal', kind='line', hue='event', style="event", col='subject', 
            col_wrap=4, linewidth=3,
           data=fmri);
```
