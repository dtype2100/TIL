```python
df = pd.read_csv("titanic_train.csv")
df.head()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
def set_age_range(a):
    if a // 10 == 0:
        return "유아"
    elif a // 10 == 1:
        return "10대"
    elif a // 10 == 2:
        return "20대"
    elif a // 10 == 3:
        return "30대"
    elif a // 10 == 4:
        return "40대"
    elif a // 10 == 5:
        return "50대"
    elif a // 10 == 6:
        return "60대"
    else:
        return "노인"
```

```python
df['Age_Range'] = df['Age'].apply(set_age_range)
df.head()
```

```python
survivied_by_ages = df.groupby('Age_Range')['Survived'].sum()
```

```python
survivied_by_ages
```

```python
survivied_by_ages = survivied_by_ages.reindex(['유아', '10대', '20대', '30대', '40대', '50대', '60대', '노인'])
survivied_by_ages
```

```python
plt.bar(survivied_by_ages.index, survivied_by_ages.values, color='g')
plt.title("나이대별 타이타닉 생존자")
plt.xlabel("나이")
plt.ylabel("생존자")
plt.show()
```

```python
survived_by_sex = df.groupby("Sex")['Survived'].sum()
```

```python
plt.bar(survived_by_sex.index, survived_by_sex.values, color='g')
plt.title("성별 타이타닉 생존자")
plt.xlabel("성")
plt.ylabel("생존자")
plt.savefig('titanic.png')
plt.show()
```
