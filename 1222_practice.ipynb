```python
def set_age_range(a):
    if a < 10:
        return "유아"
    elif a < 20:
        return "10대"
    elif a < 30:
        return "20대"
    elif a < 40:
        return "30대"
    elif a < 50:
        return "40대"
    elif a < 60:
        return "50대"
    elif a < 70:
        return "60대"
    else:
        return "노인"
    
```

```python
df['Age_Range'] = df['Age'].apply(set_age_range)
df.head()
```
