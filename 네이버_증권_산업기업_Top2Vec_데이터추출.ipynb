```python
# 2018-07-01, 2019-08-01 (미중 무역전쟁 - 일본 수출 규제 시작)
# 2019-08-01, 2020-02-01 (일본 수출 규제, 코로나19 시작)
# 2020-02-01, 2022-02-01 (코로나19, 우크라이나 사태)
# 2020-03-03, 2020-06-05 (코로나로 인한 코스피 급락 - 코로나로 인한 급등)

mask = (new_df['일자'] >= '2020-03-06') & (new_df['일자'] <= '2022-04-30')

period_df = new_df.loc[mask]
```

```python
period_df[period_df['제목'].str.contains('네이버')]
```

```python
lis_df = list(period_df['제목'])
```

```python
from top2vec import Top2Vec
```

```python
model = Top2Vec(documents=lis_df, speed="learn", workers=8, embedding_model='universal-sentence-encoder')
```

```python
model.get_num_topics()
```

```python
topic_sizes, topic_nums = model.get_topic_sizes()
```

```python
topic_words, word_scores, topic_scores, topic_nums = model.search_topics(keywords=["네이버"], num_topics=5)
```

```python
topic_words
```

```python
documents, document_scores, document_ids = model.search_documents_by_topic(topic_num=1, num_docs=85)
for doc, score, doc_id in zip(documents, document_scores, document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
doc_df = pd.DataFrame({'일자' : document_ids, '제목' : documents, 'URL' : document_scores})
```

```python
doc_df
```
