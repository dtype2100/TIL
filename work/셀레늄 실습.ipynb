```python
import requests
from bs4 import BeautifulSoupb
```

```python
url =" https://comic.naver.com/webtoon/weekdayList?week=fri"
resp = requests.get(url)
html = BeautifulSoup(resp.text, 'html.parser')
lis = html.select(".img_list >li")
```

```python
for li in lis:
    print("제목 : {}".format(li.select_one("dt").text.strip()))
    print("작가 : {}".format(li.select_one("dd.desc").text.strip()))
    print("평점 : {}".format(li.select_one("div.rating_type strong").text.strip()))
    print()
```

```python
import time

driver = webdriver.Chrome('./chromedriver')
driver.get('https://www.jungle.co.kr')
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
driver.implicitly_wait(2)

more_button = driver.find_element_by_css_selector('#existMore')
more_button.click()

time.sleep(1)

more_button = driver.find_element_by_css_selector('#existMore')
more_button.click()
```
