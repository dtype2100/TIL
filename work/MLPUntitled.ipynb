```python
import pandas as pd
df1 = pd.DataFrame({'a': ['a', 'b', 'c'], 'b': [1, 2, 3]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'a': ['a', 'b', 'd'], 'c': ['가', '나', '다']})
```

```python
df2
```

```python
pd.merge(df1, df2, how='inner', on="a") #양쪽에 동일한 Key 컬럼이 있을 경우 하나로 표현 가능
```

```python
pd.merge(df1, df2, how = 'outer', on = 'a') #두 테이블 모두 합하여 보여줌, 비어있는 셀 NaN으로 처리
```

```python
pd.merge(df1, df2, how = 'left', on ='a') #왼쪽 테이블을 기준으로 오른쪽 테이블의 내용 결합
```

```python
pd.merge(df1, df2, how = 'right', on = 'a') #오른쪽 테이블을 기준으로 외쪽 테이블을 결합
```

```python
import matplotlib.pyplot as plt

import numpy as np

random_sample = np.random.normal(loc = 10, scale = 2, size = 1000)
plt.hist(random_sample, bins = 100)
plt.show()
```

```python
import numpy as np #Q-Qplot
import pylab
import scipy.stats as stats

norm_sample = np.random.normal(loc = 20, scale = 5, size = 100)
stats.probplot(norm_sample, dist = 'norm', plot = pylab)
pylab.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import random

avg_values = []
for i in range(1, 11):
    random_sample = random.sample(range(1, 1000), 100)
    x = np.mean(random_sample)
    avg_values.append(x)
    
plt.hist(avg_values, bins = 100)
plt.show()
```
