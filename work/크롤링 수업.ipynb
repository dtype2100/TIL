```python
import requests
from bs4 import BeautifulSoup

headers = {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36"}

url = "https://news.naver.com/main/list.naver?mode=LPOD&mid=sec&oid=003"

resp = requests.get(url, headers=headers)
html = BeautifulSoup(resp.text, 'html.parser')
```

```python
lis = html.select(".list_body li")
len(lis)
```

```python
prev_day = html.select_one(".pagenavi_day > a")
prev_url = url + "&" + prev_day.attrs['href'].split("&")[-1]
print(prev_url)
```

```python
from datetime import datetime, timedelta

yesterday = datetime.today() - timedelta(1)
```

```python
prev_url = url + "&date=" + yesterday.strftime("&Y%m&d")
prev_url
```

```python
for li in lis:
    tag = li.select_one(".photo > a")
    if tag:
        href = tag.attrs["href"]
        
        resp = requests.get(href, headers=headers)
        html = BeautifulSoup(resp.text, "html.parser")
        
        title = html.select_one("#articleTitle")
        body = html.select_one("#articleBodyContents")
        
        if title:
            print("title :", title.text)
            print("body :", body.text.strip())
            print()
```
