```python
def get_month(x):
    if x[5:7] == '12' or '01' or '02':
        return 1
    elif x[5:7] == '03' or '04' or '05':
        return 2
    elif x[5:7] == '06' or '07' or '08':
        return 3
    elif x[5:7] == '09' or '10' or '11':
        return 4
```

```python
time_month = df.release_time.apply(get_month)
```
