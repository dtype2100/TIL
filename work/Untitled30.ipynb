```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [10, 14, 19, 20, 25]
plt.plot(x)
plt.show
```

```python
x=np.arange(-4.5,5,0.5)
y=x**2
plt.plot(x,y)
plt.show() 
```

```python
x=np.arange(-4.5,5,0.5)
y1=x**2
y2=x+30
y3=x**2+10
plt.plot(x,y1, color='r')
plt.plot(x,y2, c="#6ef36e")
plt.plot(x,y3, "dashed")
plt.show()
```

```python
x=np.linspace(0,10,100)
y=x**2
plt.plot(x,y,'--r')
#plt.xlim(-20,30)
#plt.ylim(-20,120)
plt.axis([-20,30,-20,120])
plt.show()
```

```python
x=np.linspace(0,10,100)
y1=x**2
y2=x**3
#
plt.plot(x,y1,'r',label='1')
plt.plot(x,y2,'g',label='2')
#x,y
plt.xlabel('x')
plt.ylabel('y') 
plt.legend(loc=2)#()
plt.legend(bbox_to_anchor=(1.04,1))#()
plt.show()
```

```python
plt.subplot(131)
plt.plot([1,2],color='r')

plt.subplot(132)
plt.plot([1,2],color='g')

plt.subplot(133)
plt.plot([1,2],color='b')

plt.show()
```

```python
plt.figure(figsize=(5,5))#

grid=plt.GridSpec(2,3)#GridSpec(,)

plt.subplot(grid[0,0])
plt.plot([1,2],color='r')
plt.subplot(grid[0,1])
plt.plot([1,2],color='g')
plt.subplot(grid[0,2])
plt.plot([1,2],color='b')
plt.subplot(grid[1,:2])
plt.plot([1,2],color='b')
plt.subplot(grid[1,2])
plt.plot([1,2],color='r')
plt.show()

```

```python
plt.subplot(grid[0,0])
plt.plot([1,2],color='r')
plt.subplot(grid[0,1])
plt.plot([1,2],color='g')
plt.subplot(grid[0,2])
plt.plot([1,2],color='b')
plt.subplot(grid[1,:2])
plt.plot([1,2],color='b')
plt.subplot(grid[1,2])
plt.plot([1,2],color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.random.rand(30)
y=np.random.rand(30)
a=np.random.rand(30)
b=np.random.rand(30)
e=np.random.rand(30)
f=np.random.rand(30)
color=np.random.rand(30)

size=100*np.random.rand(30)

plt.scatter(x,y,s=size,c=color,alpha=0.5,marker='o')
plt.scatter(a,b,s=size,c=color,alpha=0.5,marker='^')
plt.scatter(e,f,s=size,c=color,alpha=0.5,marker='+')
plt.show()
```

```python
data=[np.random.randint(1, 7) for i in range(10)]
plt.hist(data, bins=6, alpha=1)
plt.show()
```

```python
plt.bar(['10','20','30','40','50'],[np.random.randint(1,50) for i in range(5)])
```

```python
plt.barh(['10','20','30','40','50'],[np.random.randint(1,50) for i in range(5)],color='g')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.show()
```
