```python
import pandas as pd
import numpy as np
from top2vec import Top2Vec
```

```python
df['Score'] = np.nan
df
```

```python
minus_df = df[df['Topic'] == -1]
df.drop(minus_df.index, inplace=True)
```

```python
fre_df = df[df['Frequency'] == 1]
df.drop(fre_df.index, inplace=True)
```

```python
up_df = df[df['Words'].str.contains('상승')]
up_df.drop_duplicates(inplace=True)
```

```python
down_df = df[df['Words'].str.contains('하락')]
down_df.drop_duplicates(inplace=True)
```

```python
lis_up = list(up_df['Words'])
lis_down = list(down_df['Words'])
```

```python
model_down = Top2Vec(documents=lis_down, speed="learn", workers=8, embedding_model='universal-sentence-encoder')
```

```python
model_up = Top2Vec(documents=lis_up, speed="learn", workers=8, embedding_model='universal-sentence-encoder')
```

```python
model_down.get_num_topics()
```

```python
model_up.get_num_topics()
```

```python
topic_sizes, topic_nums = model_down.get_topic_sizes()
topic_words, word_scores, topic_scores, topic_nums = model_down.search_topics(keywords=["하락"], num_topics=1) # num_topics는 'model.get_num_topics()'에 나온 수와 동일하게 설정. keywords
```

```python
topic_sizes, topic_nums = model_up.get_topic_sizes()
topic_words, word_scores, topic_scores, topic_nums = model_up.search_topics(keywords=["상승"], num_topics=1) # num_topics는 'model.get_num_topics()'에 나온 수와 동일하게 설정. keywords
```

```python
down_documents, down_document_scores, down_document_ids = model_down.search_documents_by_topic(topic_num=1, num_docs=10) # 필요한 만큼 topic_num 변경, num_docs 변경
for doc, score, doc_id in zip(down_documents, down_document_scores, down_document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
up_documents, up_document_scores, up_document_ids = model_up.search_documents_by_topic(topic_num=1, num_docs=10) # 필요한 만큼 topic_num 변경, num_docs 변경
for doc, score, doc_id in zip(up_documents, up_document_scores, up_document_ids):
    print(f"Document: {doc_id}, Score: {score}")
    print("-----------")
    print(doc)
    print("-----------")
    print()
```

```python
#doc_df = pd.DataFrame({'Words' : documents}) # Top2Vec을 데이터프레임으로 만들기
down_doc_df = pd.DataFrame({'Topic' : np.nan, 'Words' : down_documents, 'Frequency': np.nan, 'Timestamp' : np.nan, 'Name' : np.nan, 'Score' : down_document_scores})
down_sum_df = pd.concat([down_doc_df, df]) # Top2Vec에서 추출한 데이터와 날짜 인덱싱한 데이터와 합치기
down_dupl_df = down_sum_df[down_sum_df['Words'].duplicated()] # Top2Vec에서 추출한 데이터와 날짜 인덱싱한 데이터에서 중복값 변수 설정
down_frame_df = pd.DataFrame({'Topic' : down_dupl_df['Topic'], 'Words' : down_dupl_df['Words'], 'Frequency' : down_dupl_df['Frequency'], 'Timestamp' : down_dupl_df['Timestamp'], 'Name' : down_dupl_df['Name'], 'Score' : down_dupl_df['Score']}) # 중복값 설정한 데이터에 datetime 순서로 데이터프레임 생성
down_sort_df = down_frame_df.sort_values(by=['Timestamp']) # 데이터프레임을 일자별로 정렬
down_reset_df = down_sort_df.reset_index()
down_f_df = pd.concat([down_doc_df, down_reset_df])
down_d_df = down_f_df[down_f_df['Words'].duplicated()]
down_r_df = down_d_df.reset_index()

#doc_df = pd.DataFrame({'Words' : documents}) # Top2Vec을 데이터프레임으로 만들기
up_doc_df = pd.DataFrame({'Topic' : np.nan, 'Words' : up_documents, 'Frequency': np.nan, 'Timestamp' : np.nan, 'Name' : np.nan, 'Score' : up_document_scores})
up_sum_df = pd.concat([up_doc_df, df]) # Top2Vec에서 추출한 데이터와 날짜 인덱싱한 데이터와 합치기
up_dupl_df = up_sum_df[up_sum_df['Words'].duplicated()] # Top2Vec에서 추출한 데이터와 날짜 인덱싱한 데이터에서 중복값 변수 설정
up_frame_df = pd.DataFrame({'Topic' : up_dupl_df['Topic'], 'Words' : up_dupl_df['Words'], 'Frequency' : up_dupl_df['Frequency'], 'Timestamp' : up_dupl_df['Timestamp'], 'Name' : up_dupl_df['Name'], 'Score' : up_dupl_df['Score']}) # 중복값 설정한 데이터에 datetime 순서로 데이터프레임 생성
up_sort_df = up_frame_df.sort_values(by=['Timestamp']) # 데이터프레임을 일자별로 정렬
up_reset_df = up_sort_df.reset_index()
up_f_df = pd.concat([up_doc_df, up_reset_df])
up_d_df = up_f_df[up_f_df['Words'].duplicated()]
up_r_df = up_d_df.reset_index()
```
