```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.__version__
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="body_mass_g", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="species", style="species", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="species", style="island", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="species", col="island", data=penguins);
```

```python
sns.set(style="whitegrid")
```

```python
sns.relplot(x="bill_length_mm", y="bill_depth_mm", hue="species", col="island", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", hue="body_mass_g", size="body_mass_g", data=penguins);
```

```python
sns.relplot(x="flipper_length_mm", y="body_mass_g", 
            hue="body_mass_g", size="body_mass_g",  sizes=(30, 100), data=penguins);
```
