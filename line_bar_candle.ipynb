```python
df = fdr.DataReader('032640', '20210901', '20220401')
```

```python
df.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Scatter(x=df.index,y=df['Close'], mode='lines'))
fig.show()
```

```python
fig = go.Figure(data=go.Scatter(x=df.index,y=df['Close'], mode='lines+markers'))
fig.show()
```

```python
from plotly.subplots import make_subplots

fig2 = make_subplots(specs=[[{"secondary_y": True}]])
fig2.add_trace(go.Scatter(x=df.index,y=df['Close'],name='Price'),secondary_y=False)
fig2.add_trace(go.Bar(x=df.index,y=df['Volume'],name='Volume'),secondary_y=True)
fig2.show()
```

```python
fig2.update_yaxes(range=[0,10000000],secondary_y=True)
fig2.update_yaxes(visible=False, secondary_y=True)
```

```python
fig3 = make_subplots(specs=[[{"secondary_y": True}]])
fig3.add_trace(go.Candlestick(x=df.index,
                              open=df['Open'],
                              high=df['High'],
                              low=df['Low'],
                              close=df['Close'],
                              increasing_line_color = 'red',
                              decreasing_line_color = 'blue'
                              ))
```

```python
fig3.add_trace(go.Bar(x=df.index, y=df['Volume'], name='Volume'),secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)
```

```python
fig3.update_yaxes(range=[0,10000000],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
```

```python
fig3.add_trace(go.Scatter(x=df.index,y=df['Close'].rolling(window=5).mean(),marker_color='green',name='5 Day MA'))
fig3.add_trace(go.Scatter(x=df.index,y=df['Close'].rolling(window=20).mean(),marker_color='red',name='20 Day MA'))
fig3.add_trace(go.Scatter(x=df.index,y=df['Close'].rolling(window=60).mean(),marker_color='orange',name='60 Day MA'))
fig3.add_trace(go.Scatter(x=df.index,y=df['Close'].rolling(window=120).mean(),marker_color='purple',name='120 Day MA'))
fig3.add_trace(go.Bar(x=df.index, y=df['Volume'], name='Volume'),secondary_y=True)
fig3.update_layout(title={'text':'LGU', 'x':0.5})
fig3.update_yaxes(range=[0,1000000000],secondary_y=True)
fig3.update_yaxes(visible=False, secondary_y=True)
fig3.update_layout(xaxis_rangeslider_visible=False)  #hide range slider
fig3.show()
```
